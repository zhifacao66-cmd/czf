## 1. 数据加载与概览

```python
df = pd.read_csv('../data/user_behavior.csv', parse_dates=['timestamp'])
print(f'数据量: {len(df):,} 条')
print(f'时间范围: {df["timestamp"].min()} ~ {df["timestamp"].max()}')
print(f'用户数: {df["user_id"].nunique():,}')
print(f'商品数: {df["item_id"].nunique():,}')
df.head(10)
```

```python
# 缺失值与数据类型
df.info()
print(f'\n缺失值: {df.isnull().sum().sum()}')
print(f'重复行: {df.duplicated().sum()}')
```

## 2. 用户增长趋势

```python
df['date'] = df['timestamp'].dt.date
dau = df.groupby('date')['user_id'].nunique().reset_index()
dau.columns = ['date', 'dau']
dau['date'] = pd.to_datetime(dau['date'])

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(dau['date'], dau['dau'], color='#2196F3', linewidth=1.2)
ax.fill_between(dau['date'], dau['dau'], alpha=0.15, color='#2196F3')
ax.axhline(dau['dau'].mean(), color='red', linestyle='--', alpha=0.7)
ax.set_title('日活跃用户趋势 (DAU)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
print(f'平均日活: {dau["dau"].mean():.0f}')
```

## 3. 转化漏斗分析

```python
funnel = df.groupby('behavior_type')['user_id'].nunique()
stages = ['pv', 'fav', 'cart', 'buy', 'pay']
labels = ['浏览', '收藏', '加购', '下单', '支付']
counts = [funnel.get(s, 0) for s in stages]

funnel_df = pd.DataFrame({'阶段': labels, '用户数': counts})
funnel_df['总体转化率'] = (funnel_df['用户数'] / funnel_df['用户数'].iloc[0] * 100).round(2)
funnel_df['环节转化率'] = [100.0] + [
    round(funnel_df['用户数'].iloc[i] / funnel_df['用户数'].iloc[i-1] * 100, 2)
    for i in range(1, len(funnel_df))
]
funnel_df
```

## 4. 用户留存分析

```python
first = df.groupby('user_id')['date'].min().reset_index()
first.columns = ['user_id', 'first_date']
first['first_date'] = pd.to_datetime(first['first_date'])
df_m = df.merge(first, on='user_id')
df_m['day_offset'] = (df_m['date'] - df_m['first_date']).dt.days

total_u = df['user_id'].nunique()
for d in [1, 3, 7, 14, 30]:
    r = df_m[df_m['day_offset'] == d]['user_id'].nunique()
    print(f'Day-{d:2d} 留存: {r / total_u * 100:.1f}%')
```

## 5. RFM 用户分层

```python
pay = df[df['behavior_type'] == 'pay']
now = df['timestamp'].max()
rfm = pay.groupby('user_id').agg(
    last_pay=('timestamp', 'max'),
    frequency=('behavior_id', 'count'),
    monetary=('price', 'sum')
).reset_index()
rfm['recency'] = (now - rfm['last_pay']).dt.days

rfm['R'] = pd.qcut(rfm['recency'], q=4, labels=[4,3,2,1]).astype(int)
rfm['F'] = pd.qcut(rfm['frequency'].rank(method='first'), q=4, labels=[1,2,3,4]).astype(int)
rfm['M'] = pd.qcut(rfm['monetary'].rank(method='first'), q=4, labels=[1,2,3,4]).astype(int)
rfm['score'] = rfm['R'] + rfm['F'] + rfm['M']

def label(s):
    if s >= 10: return '高价值用户'
    elif s >= 8: return '重要发展用户'
    elif s >= 6: return '重要保持用户'
    elif s >= 4: return '一般用户'
    else: return '流失用户'

rfm['segment'] = rfm['score'].apply(label)
rfm['segment'].value_counts()
```
